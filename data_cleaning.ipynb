```python
# instruction1
instruction = 'READ_REG 0x06[1:2]_0x0'
# instruction = 'READ_REG 0x06_0x0'
# instruction = 'READ_REG 0x06[4]'
def Register_data__extract(instrunction: str):
    pattern1 = re.compile(r"\b0[xX]([0-9a-fA-F]+)+\[(.*?)\]+[_]+0[xX]([0-9a-fA-F]+)\b") # sample data 0x00[1:2]_0x0
    pattern2 = re.compile(r"\b0[xX]([0-9a-fA-F]+)+[_]+0[xX]([0-9a-fA-F]+)\b") # sample data 0x00_0x0
    pattern3 = re.compile(r"[xX]([0-9a-fA-F]+)+\[(.*?)\]") # sample data 0x00[1:2]
    # parse the data for the bit information and the register information 
    if re.search(pattern1, instrunction):
        instrunction = re.findall(pattern1, instrunction)[0]
        # parse only register information and the register value 
    elif re.search(pattern2, instrunction): 
       instrunction = re.findall(pattern2, instrunction)[0]
       
    elif re.search(pattern3, instrunction): 
       instrunction = re.findall(pattern3, instrunction)[0]
    
    else: 
        instruction = None 

    return instrunction
Register_data__extract(instrunction=instruction)
```

##### Force Signal Extraction 

```python
# Force signal extraction 
instruction = 'Force__VBIAS__5V'
def Force_signal__extract(instruction: str):
    pattern = re.compile(r'Force__+([0-9a-zA-Z]+)__+([0-9a-zA-Z]+)') # extract the signal and the signal value 
    if re.search(pattern, instruction):
        signal = re.findall(pattern, instruction)[0]
        instruction = {signal[0]: signal[1]}
        # print(signal)
    return instruction
Force_signal__extract(instruction)
```

##### Delay Extraction 

```python
# Delay signal extraction 
instruction = 'Wait__delay__250us'
def Delay__extract(instruction: str):
    pattern = re.compile(r'Wait__delay__+([0-9a-zA-Z]+)') # extract the signal and the signal value 
    if re.search(pattern, instruction):
        delay = re.findall(pattern, instruction)[0]
        instruction = delay
        # print(signal)
    return instruction
Delay__extract(instruction)
```
